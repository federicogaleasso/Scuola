##3. Classificazione Automatica

> L'obiettivo di questo algoritmo è quello di ***trovare l'iperpiano (o linea in un piano a 2D) che divida al meglio un set di dati in due classi***.

***Qual è l'iperpiano migliore?***

> L'iperpiano che abbia la distanza massima tra tutti i punti e che giace ***`in mezzo alla delimitazione del margine`*** (ossia lo spazio che separa l'iperpiano dai punti più vicini delle due classi).
